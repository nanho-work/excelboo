```python
df.isna().sum()
# 접수일 기준으로 연도와 월 컬럼 분리 생성
df['접수년'] = df['접수일'].dt.year
df['월'] = df['접수일'].dt.month
recent_months = df["월"].sort_values().unique()[-2:]
filtered_df = df[df["월"].isin(recent_months)]
print(df["월"].unique())
```

```python
df["월"].unique()
```

```python
# 가정: filtered_df가 이전에 처리된 상태라고 가정
df = filtered_df.fillna("").infer_objects(copy=False)
df.columns = df.columns.str.strip()

# 필수 컬럼 확인
if "월" not in df.columns or "카드사" not in df.columns or "가맹점명" not in df.columns:
    print("필수 컬럼이 없습니다.")
else:
    df["카드사"] = df["카드사"].astype(str).str.strip()
    df["가맹점명"] = df["가맹점명"].astype(str).str.strip()

    # 민원건수 집계
    grouped = df.groupby(["월", "가맹점명", "카드사"]).size().reset_index(name="민원건수")

    # 피벗 테이블 생성
    pivot = grouped.pivot_table(
        index=["가맹점명", "카드사"],
        columns="월",
        values="민원건수",
        fill_value=0
    ).sort_index(axis=1)

    # ▶️ 사용자 선택 월을 기준으로 전월 자동 계산
    selected_month = 8  # 예: 7월을 선택한 경우
    pre_month = selected_month - 1

    if pre_month not in pivot.columns or selected_month not in pivot.columns:
        print(f"비교 가능한 월({pre_month}, {selected_month}) 데이터가 부족합니다.")
    else:
        pivot = pivot.reset_index()
        merchants = pivot["가맹점명"].unique()
        card_companies = pivot["카드사"].unique()
        result_df = pd.DataFrame({"가맹점명": merchants})

        for card in card_companies:
            card_df = pivot[pivot["카드사"] == card]
            pre_counts = card_df.set_index("가맹점명").get(pre_month, pd.Series(dtype=int))
            cur_counts = card_df.set_index("가맹점명").get(selected_month, pd.Series(dtype=int))

            pre_counts = pre_counts.reindex(merchants, fill_value=0)
            cur_counts = cur_counts.reindex(merchants, fill_value=0)

            # 증감률 계산
            rate_display = []
            for i in range(len(pre_counts)):
                pre = pre_counts.iloc[i]
                cur = cur_counts.iloc[i]
                if pre == 0:
                    if cur == 0:
                        rate_display.append("0.0%")
                    else:
                        rate_display.append("신규")
                else:
                    diff_rate = ((cur - pre) / pre) * 100
                    rate_display.append(f"{diff_rate:.1f}%")

            result_df[f"{card}_전월"] = pre_counts.values
            result_df[f"{card}_당월"] = cur_counts.values
            result_df[f"{card}_증감률"] = rate_display

        # 결과 출력
        pd.set_option('display.float_format', '{:.1f}'.format)
        display(result_df)
```
